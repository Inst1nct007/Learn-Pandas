<h3>Create a Dataframe from a Dictionary of Lists</h3>

```python
d1 = {
    "Name": ['Ram', 'Shyam', 'Raj', 'Ajoy'],
    "Age": [18, 25, 92, 23]
}

df1 = pd.DataFrame(d1)
df1
```

<h3>Create a Dataframe from a List of Dictionaries</h3>

<h4>Used to make dataframes from APIs</h4>

```python
d2 = [
    {"firstName": "John", "lastName": "Doe", "city": "Pune"},
    {"firstName": "Jane", "lastName": "Doe", "city": "Bangalore"}
]

df2 = pd.DataFrame(d2)
df2
```

<h3>Create a Dataframe from a List of Lists</h3>

```python
d3 = [
    [1, "Raj", "Sales Manager", 10000],
    [2, "Utpal", "Doctor", 50000],
    [3, "Piku", "Software Developer", 1000],
    [4, "Bikas", "Mechanical Engineer", 10000],
    [5, "Shiv", "UI & UX Designer", 20000],
    [6, "John", "Painter", 100000],
    [7, "Jane", "Carpenter", 40000],
    [8, "Dave", "Civil Engineer", 90000],
    [9, "Marc", "Cleaner", 20000],
    [10, "Raj", "Animator", 90000],
    [11, "Dave", "Doctor", 150000],
    [12, "Mina", "Software Developer", 250000],
]

df3 = pd.DataFrame(d3, columns=["ID", "Name", "Job", "Salary"])
df3
```

<h3>Create a Dataframe from a Dictionary of Dictionaries</h3>

```python
d4 = {
    "Name": {
        "0": "Mohan",
        "1": "Ayush"
    },
    "Age": {
        "0": 26,
        "1": 73
    },
}

df4 = pd.DataFrame(d4)
df4
```

<h3>Read a CSV File</h3>

```python
df5 = pd.read_csv('file.csv', index_col='ID')
df5
```

<h3>Use loc[ ] to Filter Rows</h3>

```python
df5.loc[5: 13: 3, :] # : means all the columns
```

<h3>Use loc[ ] to Filter Particular Columns ( Name , Salary )</h3>

```python
df5.loc[:,["Name", "Salary"]] # : means all the rows
```

<h3>Use loc[ ] to Filter from One Column to Another Column ( Name to Salary )</h3>

```python
df5.loc[:,"Name": "Salary"]
```

<h3>Sort DataFrame in Descending Order of Salary then Name</h3>

<h6>If two people have the same salary they will be sorted according to their names (Desc Order)</h6>

```python
df5.sort_values(by=["Salary", "Name"], ascending=False)
```
